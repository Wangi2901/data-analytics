**BOX-PLOT**

```python
Q1 = apps['Maximum Installs'].quantile(0.25)
Q2 = apps['Maximum Installs'].quantile(0.75)
upper_threshold = Q2 + 1.5 * (Q2 - Q1)
print("Schwellenwert: " + str(upper_threshold) + '\n')

apps_18300 = apps[apps['Maximum Installs'] <= 18300]
fig, ax = plt.subplots(ncols=2, figsize=(10,10))
ax[0].boxplot(apps['Maximum Installs'], labels=['Anzahl Installationen'])
ax[1].boxplot(apps_18300['Maximum Installs'], labels=['Anzahl Installationen'])
plt.show()
```

**SCATTER-PLOT**

**AUFGABE:**


Statistische Zusammenfassung von mindestens einer Spalte --> Size

```python
# Entferne Strings
# size = apps['Size'].str.replace(pat='Varies with device', repl='', regex=False).str.replace(pat='M', repl='', regex=False).str.replace(pat='k', repl='', regex=False)
size = apps.Size.str.extract(r'([\d]+)', expand=False)

# String in float umwandeln
size = pd.to_numeric(size, errors='coerce')

print(f'Durchschnittliche Größe der Apps: {size.mean():.2f}')
print(f'Preise:')
print(f'  Min:        {size.min():5.2f}')
print(f'  1. Quartil: {size.quantile(0.25):5.2f}')
print(f'  Median:     {size.quantile(0.50):5.2f}')
print(f'  3. Quartil: {size.quantile(0.75):5.2f}')
print(f'  Max:        {size.max():5.2f}')
print(f'Größe aller Apps summiert in GB: {size.sum() / 1000.0:.2f}')
```
